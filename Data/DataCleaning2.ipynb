# Clean Classification column

## Extract salary info fromit and make a new column 'Salary'

```python
def apply_saraly(a):
    if 'classification:' in a:
        return None
    else:
        return a
```

```python
data['Salary'] = data.Classification.apply(apply_saraly)
data.head(20)
```

## Clean saraly info in 'Classsification' column

```python
def clean_salary(a):
    if 'classification:' in a:
        return a
    else:
        return None
```

```python
data.Classification = data.Classification.apply(clean_salary)
data.head(20)
```

## Split  'Classification' into 'Subclassification' and 'Classification' 

```python
data[['Classification', 'Subclassification']] = data.Classification.str.split('subClassification:', expand=True)
```

```python
data.head(20)
```

```python
data.Classification = data.Classification.str.replace('classification:','')
data.head(20)
```

## Deduplicate Subclassification

```python
data.Subclassification.head(20)
```

```python
def deduplication(x):
    x = x.strip()
    if x == 'None':
        return None
    return re.match(r'(.*)\1', x).group(1)
```

```python
data.Subclassification = data.Subclassification.astype('str')
data.Subclassification.dtype
```

```python
data.Subclassification = data.Subclassification.apply(deduplication)
data.head(20)
```

## Deduplicate Classification

```python
data.Classification = data.Classification.astype('str')
data.Classification = data.Classification.apply(deduplication)
data.head(20)
```

# Clean 'Time Posted'

```python
# 不是时间的一律为0
def apply_TP(x):
    if re.match(r'^(\d[a-z]\s)',x):
        return re.match(r'^(\d[a-z])\s',x).group(1)
    else:
        return 0
```

```python
x = '7d ago,at,Private Advertiser'
apply_TP(x)
```

```python
y = 'Featured,at,Private Advertiser'
apply_TP(y)
```

```python
data['Time Posted'] = data['Time Posted'].astype('str')
data['Time Posted'] = data['Time Posted'].apply(apply_TP)
data.head(20)
```

```python
a='234d'
a[:-1]
```

```python
def date_transfer(x):
    if x == '0' or 'h' in x:
        return 0
    y = int(x[:-1])
    if 'd' in x:
        return y
    # 假定一半月份是大月另一半是小月
    if 'm' in x:
        return y // 2 * 31 + (y - y // 2) * 30
```

```python
x = '9m'
date_transfer(x)
```

```python
data['Time Posted'] = data['Time Posted'].astype('str')
data['Time Posted'] = data['Time Posted'].apply(date_transfer)
data.head(20)
```
