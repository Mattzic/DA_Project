```python
# ÂÖàÂà§Êñ≠x‰∏≠ÊòØÂê¶Âê´ÊúâÊï∞Â≠ó
# Â¶ÇÊûúÊúâÂê´ÊúâÊï∞Â≠óÔºåÂú®Âà§Êñ≠ÊòØÂê¶ÊòØÁôæÂàÜÊï∞
# Â¶ÇÊûúÊòØÁôæÂàÜÊï∞ÔºåÂàôËØ¥Êòéx‰∏çÊòØÂ∑•ËµÑÊï∞ÔºåÂ¶Ç "8% holiday pay"Ôºå‰∏çËøõË°åËΩ¨Âåñ
# Â¶ÇÊûúÊï∞Â≠óÔºå‰∏îË°®Á§∫Ëñ™Ê∞¥Êï∞ÈáèÔºåÂ¶ÇÊûúÂ∞è‰∫é500ÔºåÂàôËÆ§‰∏∫ÊòØÂ∞èÊó∂Êñ∞
# ËΩ¨Êç¢Ôºö8Â∞èÊó∂Ôºå20Â§©Ôºå10‰∏™Êúà
# ÊâÄÊúâÂåÖÂê´Ë°®Á§∫Ëñ™Ê∞¥Êï∞Â≠óÁöÑxÂÖ®ÈÉ®ËΩ¨Âåñ‰∏∫Âπ¥Ëñ™Êï∞Â≠óÔºåÊ≤°ÊúâÂ≠óÁ¨¶Ôºå‰æã"NZD75000 per annum" -> "75000.0"; "NZD25" -> "40000.0"
def to_annual(x):
    matchObj = re.match('[A-Z]*(\d+\.*\d*)', x)
    if matchObj:
        if re.match('[A-Z]*\d+\.*\d*\s*\%', x):
            pass
        else:
            x = float(matchObj.group(1))
            if x < 999:
                x = x * 8 * 20 * 10
    return x
         
```

```python
data['Lo Salary'] = data['Lo Salary'].apply(to_annual)
data[['Lo Salary', 'Hi Salary']].head(300)
```

## Fill all nan's with some artificially made data

```python
# Êúâ‰∫õÊó∂ÂÄôÂéüSalaryÂàóÊòØÁ±ª‰ºº "Up to $23 p.h. + + 8% Holiday Pay"ÁöÑÂ≠óÁ¨¶‰∏≤Ôºå
# ËøôÊÑèÂë≥ÁùÄLo Salary ÂèñÊï∞ÁöÑÊó∂ÂÄô‰∏çÂèØ‰ª•ÊØîHi Salary È´ò
data['Lo Salary'] = data['Lo Salary'].astype('str')
data[['Lo Salary', 'Hi Salary']][data['Lo Salary'] == 'Up ']
```

```python
# Êü•ÁúãÂè™ÊúâHiÊúâÊï∞ÊçÆÔºå‰ΩÜLoÊ≤°ÊúâÊï∞ÊçÆÁöÑË°å
data[['Lo Salary', 'Hi Salary']][(data['Lo Salary'] == 'nan') & ~(data['Hi Salary'].isnull())]
```

```python
# Ëé∑ÂèñLo Salary ÁöÑÊúÄÂÄº
df1 = data[data['Lo Salary'].str.contains(r'\d', regex=True, na=False)]['Lo Salary']
```

```python
def set0(x):
    if re.match(r'^\d\d+',x):
        return x
    elif re.match(r'.*\s\d\d+',x):
        return re.match(r'.*\s(\d\d+)',x).group(1)
    else:
        return 0
    
df1 = df1.apply(set0)
```

```python
df1 = df1.drop([201,662,691,1945])
df1
```

```python
df1 = df1.astype('float')
print(df1.min())
print(df1.max())
```

```python
# ÈúÄË¶ÅÂú®ËåÉÂõ¥‰πãÂÜÖÈöèÊú∫ÂèñÊï∞
from numpy.random import default_rng

def set_value(x):
    rng = default_rng()
    if x == 'nan':
        x = rng.integers(29600, 112000, endpoint=True)
    return x
```

```python
data['Lo Salary'] = data['Lo Salary'].apply(set_value)
data['Lo Salary'].head(200)
```
