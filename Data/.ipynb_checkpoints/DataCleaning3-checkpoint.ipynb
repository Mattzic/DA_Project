```python
# ÂÖàÂà§Êñ≠x‰∏≠ÊòØÂê¶Âê´ÊúâÊï∞Â≠ó
# Â¶ÇÊûúÊúâÂê´ÊúâÊï∞Â≠óÔºåÂú®Âà§Êñ≠ÊòØÂê¶ÊòØÁôæÂàÜÊï∞
# Â¶ÇÊûúÊòØÁôæÂàÜÊï∞ÔºåÂàôËØ¥Êòéx‰∏çÊòØÂ∑•ËµÑÊï∞ÔºåÂ¶Ç "8% holiday pay"Ôºå‰∏çËøõË°åËΩ¨Âåñ
# Â¶ÇÊûúÊï∞Â≠óÔºå‰∏îË°®Á§∫Ëñ™Ê∞¥Êï∞ÈáèÔºåÂ¶ÇÊûúÂ∞è‰∫é500ÔºåÂàôËÆ§‰∏∫ÊòØÂ∞èÊó∂Êñ∞
# ËΩ¨Êç¢Ôºö8Â∞èÊó∂Ôºå20Â§©Ôºå10‰∏™Êúà
# ÊâÄÊúâÂåÖÂê´Ë°®Á§∫Ëñ™Ê∞¥Êï∞Â≠óÁöÑxÂÖ®ÈÉ®ËΩ¨Âåñ‰∏∫Âπ¥Ëñ™Êï∞Â≠óÔºåÊ≤°ÊúâÂ≠óÁ¨¶Ôºå‰æã"NZD75000 per annum" -> "75000.0"; "NZD25" -> "40000.0"
def to_annual(x):
    matchObj = re.match('[A-Z]*[a-z]*\s*(\d+\.*\d*)', x)
    if matchObj:
        if re.match('[A-Z]*\d+\.*\d*\s*\%', x):
            pass
        else:
            x = float(matchObj.group(1))
            if x < 60:
                x = x * 8 * 20 * 10
            elif x< 250:
                x = x * 1000
            else:
                pass
    return x
         
```

```python
data['Lo Salary'] = data['Lo Salary'].apply(to_annual)
```

```python
def Clean_Convert(x):
    matchObj = re.match('[^0-9]*(\d+\.*\d*)\s*[a-z]*',x)
    if matchObj:
        if re.match('.*\d+\.*\d*\s*\%', x):
            pass
        x = float(matchObj.group(1))
        if x < 500:
            x = x * 8 * 20 * 10
    return x


data['Hi Salary'] = data['Hi Salary'].apply(Clean_Convert)
```

### Change 'None' to some number which satisfies some constrains

```python
df1 = df1.astype('float')
floor = df1.min()
ceil = df1.max()
print(floor)
print(ceil)
```

```python
data[['Lo Salary', 'Hi Salary']].head(500)
```

```python
data['Hi Salary'] = data['Hi Salary'].astype('str')
data['Lo Salary'] = data['Lo Salary'].astype('str')
rng = default_rng()

# less than ceil, greater than max(Lo Salary, floor)
for i, v in enumerate(data['Hi Salary'].values):
    v = v.strip()
    if v == 'None':
        if re.match('^\d+\.*\d*$', data['Lo Salary'][i]):
            Lo = float(data['Lo Salary'][i])
            floor = max(floor, Lo)
            num = rng.integers(max(floor,float(data['Lo Salary'][i])), ceil)        
        # if it is str in Lo Salary like 'Good salary' or 'bonus plus medical insurance', do noting
        elif re.match('.*[a-z]+\s*$', data['Lo Salary'][i]):
            continue
        else:
            num = rng.integers(floor, ceil)
        data.loc[i, 'Hi Salary'] = num

data[['Salary','Lo Salary', 'Hi Salary']].head(500)
```

```python
def find_max(x):
    res = 0
    for v in x.values:
        if re.match(r'.*[0-9]+$', v):
            v = float(v)
#             print(v)
            if v > res:
                res = v
    return res
max = find_max(df1)
print(max)
```
