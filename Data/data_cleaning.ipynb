## Extract salary from Classification and make them a new column named 'Salary'. For rows of no salary info, filled it with 'Not Specified'

#### Check the dtype of Classification

```python
data.Classification.dtype
```

#### Extracting cla info from Cla, put them in a new df

```python
s = data.Classification.str.extract(r'(^class.*)', expand=False)
df1 = data[data.Classification == s.values]
df1
```

#### Rename the Classifaction column of data as Salary 

```python
data = data.rename({'Classification': 'Salary'}, axis=1)
data.head(10)
print(len(data))
```

```python
data[data['Job Title'] == 'Card Fulfillment Officer']
```

#### Join two table on keys[Job Title,  	Link,	Orgnisation,	Location,	Time Posted]

```python
data = pd.merge(data, df1, how='left', on=['Job Title', 'Link', 'Orgnisation', 'Location', 'Time Posted'])
data.head(10)
```

```python
print(len(data))
```

```python
data[data['Job Title'] == 'Card Fulfillment Officer']
```

#### Replace the clas info in columns Salary with 'Not specified' 

```python
data.Salary.replace(to_replace=r'^class.*', value='Not specified', regex=True, inplace=True)
```
